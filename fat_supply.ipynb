```python
# Top 10 obese countries
top_obesed = df.sort_values(by = "Obesity", ascending = False)[["Country", "Obesity"]].head(10)
top_obesed
```

```python
# Top 10 Undernourished countries
malnourished = df.sort_values(by = "Undernourished", ascending=False)[["Country", "Undernourished"]][:10]
malnourished
```

```python
df["Undernourished"].value_counts()
```

```python
# Top 10 countries consuming fats through alcoholic beverages
fat_from_alcohol = df.sort_values(by = "Alcoholic Beverages", ascending=False)[["Country", "Alcoholic Beverages"]][:10]
fat_from_alcohol
```

```python
# Top 10 countries consuming fats through animal products
fat_from_animal_products = df.sort_values(by = "Animal Products", ascending=False)[["Country", "Animal Products"]][:10]
fat_from_animal_products
```

```python
# Top 10 countries consuming fats through animal fatss
fat_from_animal_fats = df.sort_values(by = "Animal fats", ascending=False)[["Country", "Animal fats"]][:10]
fat_from_animal_fats
```

```python
# Top 10 countries consuming fats through oil crops
fat_from_Oilcrops = df.sort_values(by = "Oilcrops", ascending=False)[["Country", "Oilcrops"]][:10]
fat_from_Oilcrops
```

```python
fig, (ax1, ax2, ax3, ax4) = plt.subplots(nrows = 4, figsize = (15,20))
ax1.bar(top_obesed["Country"], top_obesed["Obesity"])
ax1.set_ylabel("Obesity")
ax2.bar(fat_from_animal_products["Country"], fat_from_animal_products["Animal Products"], color = "salmon")
ax2.set_ylabel("Fat from Animal Products")
ax3.bar(fat_from_animal_fats["Country"], fat_from_animal_fats["Animal fats"], color = "lightblue")
ax3.set_ylabel("Animal Fats Consumption")
ax4.bar(fat_from_Oilcrops["Country"], fat_from_Oilcrops["Oilcrops"], color = "orange")
ax4.set_ylabel("Fat from Oilcrops")
fig.suptitle("Fat Consumptions Through various sources")
plt.show()
```

```python
# Nepal's Data
Nepal = df[df["Country"] == "Nepal"]
Nepal
```

```python
Covid_status_npl = {"Confirmed": Nepal["Confirmed"],
                    "Active": Nepal["Active"],
               "Recovered": Nepal["Recovered"],
               "Death": Nepal["Deaths"]}
```

```python
nepl_covid_df = pd.DataFrame(Covid_status_npl)
```

```python
nepl_covid_df.T.plot(kind = "bar", figsize = (10,6))
plt.xticks(rotation = 0)
plt.legend(["Percentage of people of Nepal"]);
```

```python
Nepal_fat_consumptions = Nepal.drop(["Population","Country","Confirmed", "Active", "Recovered", "Deaths","Unit (all except Population)"], axis = 1)
```

```python
Nepal_fat_consumptions.plot(kind = "bar", figsize = (18,15));
plt.xticks(rotation = 0)
plt.title("Nepal's Fat intake")
```
